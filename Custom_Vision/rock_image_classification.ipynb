Create a new Python application

```python
# Prediction module from Azure Custom Vision API 
from azure.cognitiveservices.vision.customvision.prediction import CustomVisionPredictionClient
# OpenAPI Authentication
from msrest.authentication import ApiKeyCredentials
# Visualization
from matplotlib import pyplot as plt
# Image handliing
from PIL import Image, ImageDraw, ImageFont
# Mathematical and scientific arithmetics
import numpy as np
# file handling
import os
```

Create variables for your resource's Azure endpoint and keys.

```python
prediction_endpoint = ""
prediction_key = ""
project_id = ""
model_name = ""
```

Authenticate the client

```python
credentials = ApiKeyCredentials(in_headers= {"Prediction-key": prediction_key})
predictor = CustomVisionPredictionClient(endpoint = prediction_endpoint, credentials = credentials)
```

Load image

```python
image_file = "data/Basalt_Cristobalite_s69-45540.jpg"
image = Image.open(image_file)
```

Detect objects in the test image

```python
with open(image_file, mode = "rb") as image_data:
    results = predictor.classify_image(project_id, model_name, image_data)

plt.imshow(image)

for prediction in results.predictions:
    print(f"Tag: {prediction.tag_name}\t Probability: {prediction.probability:.2f}")
```
